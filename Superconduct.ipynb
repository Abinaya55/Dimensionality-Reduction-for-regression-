```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import time
df=pd.read_csv("Chemical.csv")
#print(df)
X=df.iloc[:,0:80]
#print(X)
y=df.iloc[:,81]
#print(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=6)

#Linear model prediction
start=time.time()
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
end = time.time()
print("Execution time-linear regression: ",end - start)
y_pred = model.predict(X_test)
print("R2 Score-Linear Model",r2_score(y_test, y_pred))

#Ridge model prediction
start=time.time()
model = Ridge()
model.fit(X_train, y_train)
end = time.time()
print("Execution time-Ridge regression: ",end - start)
y_pred = model.predict(X_test)
print("R2 Score-Ridge Model",r2_score(y_test, y_pred))

#lasso model prediction
start=time.time()
model = Lasso()
model.fit(X_train, y_train)
end = time.time()
print("Execution time-lasso regression: ",end - start)
y_pred = model.predict(X_test)
print("R2 Score-lasso Model",r2_score(y_test, y_pred))


#ElasticNet model prediction
start=time.time()
model = ElasticNet()
model.fit(X_train, y_train)
end = time.time()
print("Execution time-ElasticNet regression: ",end - start)
y_pred = model.predict(X_test)
print("R2 Score-ElasticNet Model",r2_score(y_test, y_pred))

#GradientBoostingRegressor model prediction
start=time.time()
model = GradientBoostingRegressor()
model.fit(X_train, y_train)
end = time.time()
print("Execution time-GradientBoostingRegressor : ",end - start)
y_pred = model.predict(X_test)
print("R2 Score-GradientBoostingRegressor",r2_score(y_test, y_pred))

# RandomForestRegressor model prediction
start=time.time()
model = RandomForestRegressor(max_depth=5, random_state=0)
model.fit(X_train, y_train)
end = time.time()
print("Execution time-random forest regression: ",end - start)
y_pred = model.predict(X_test)
print("R2 Score-Random forest",r2_score(y_test, y_pred))

```

```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.decomposition import PCA
import time
df=pd.read_csv("Chemical.csv")
#print(df)
X=df.iloc[:,0:80]
#print(X)
y=df.iloc[:,81]
#print(y)


#Uisng Dimensionality Reduction--PCA 
start=time.time()
pca = PCA(n_components=5)
pca.fit(X)
trans=pca.transform(X)
end = time.time()
print("Execution for transformation PCA : ",end - start)

X_train, X_test, y_train, y_test = train_test_split(trans, y, test_size=0.2, random_state=6)

#Linear model prediction
start=time.time()
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
end = time.time()
print("Execution time-linear regression: ",end - start)
y_pred = model.predict(X_test)
print("R2 Score-Linear Model",r2_score(y_test, y_pred))

#Ridge model prediction
start=time.time()
model = Ridge()
model.fit(X_train, y_train)
end = time.time()
print("Execution time-Ridge regression: ",end - start)
y_pred = model.predict(X_test)
print("R2 Score-Ridge Model",r2_score(y_test, y_pred))

#lasso model prediction
start=time.time()
model = Lasso()
model.fit(X_train, y_train)
end = time.time()
print("Execution time-lasso regression: ",end - start)
y_pred = model.predict(X_test)
print("R2 Score-lasso Model",r2_score(y_test, y_pred))


#ElasticNet model prediction
start=time.time()
model = ElasticNet()
model.fit(X_train, y_train)
end = time.time()
print("Execution time-ElasticNet regression: ",end - start)
y_pred = model.predict(X_test)
print("R2 Score-ElasticNet Model",r2_score(y_test, y_pred))

#GradientBoostingRegressor model prediction
start=time.time()
model = GradientBoostingRegressor()
model.fit(X_train, y_train)
end = time.time()
print("Execution time-GradientBoostingRegressor : ",end - start)
y_pred = model.predict(X_test)
print("R2 Score-GradientBoostingRegressor",r2_score(y_test, y_pred))

# RandomForestRegressor model prediction
start=time.time()
model = RandomForestRegressor(max_depth=5, random_state=0)
model.fit(X_train, y_train)
end = time.time()
print("Execution time-random forest regression: ",end - start)
y_pred = model.predict(X_test)
print("R2 Score-Random forest",r2_score(y_test, y_pred))
```

```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn import datasets, cluster
import time

df=pd.read_csv("Chemical.csv")
#print(df)
X=df.iloc[:,0:80]
#print(X)
y=df.iloc[:,81]
#y data type is changed to int
y=y.astype('int')
#print(y)


#Uisng Dimensionality Reduction--FA 
start=time.time()
agglo = cluster.FeatureAgglomeration(n_clusters=4)
agglo.fit(X)
X_reduced = agglo.transform(X)
X_reduced.shape
end = time.time()
print("Execution for transformation FA: ",end - start)

X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size=0.2, random_state=6)

#Linear model prediction
start=time.time()
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
end = time.time()
print("Execution time-linear regression: ",end - start)
y_pred = model.predict(X_test)
print("R2 Score-Linear Model",r2_score(y_test, y_pred))

#Ridge model prediction
start=time.time()
model = Ridge()
model.fit(X_train, y_train)
end = time.time()
print("Execution time-Ridge regression: ",end - start)
y_pred = model.predict(X_test)
print("R2 Score-Ridge Model",r2_score(y_test, y_pred))

#lasso model prediction
start=time.time()
model = Lasso()
model.fit(X_train, y_train)
end = time.time()
print("Execution time-lasso regression: ",end - start)
y_pred = model.predict(X_test)
print("R2 Score-lasso Model",r2_score(y_test, y_pred))


#ElasticNet model prediction
start=time.time()
model = ElasticNet()
model.fit(X_train, y_train)
end = time.time()
print("Execution time-ElasticNet regression: ",end - start)
y_pred = model.predict(X_test)
print("R2 Score-ElasticNet Model",r2_score(y_test, y_pred))

#GradientBoostingRegressor model prediction
start=time.time()
model = GradientBoostingRegressor()
model.fit(X_train, y_train)
end = time.time()
print("Execution time-GradientBoostingRegressor : ",end - start)
y_pred = model.predict(X_test)
print("R2 Score-GradientBoostingRegressor",r2_score(y_test, y_pred))

# RandomForestRegressor model prediction
start=time.time()
model = RandomForestRegressor(max_depth=5, random_state=0)
model.fit(X_train, y_train)
end = time.time()
print("Execution time-random forest regression: ",end - start)
y_pred = model.predict(X_test)
print("R2 Score-Random forest",r2_score(y_test, y_pred))
```
